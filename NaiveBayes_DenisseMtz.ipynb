# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
import pandas as pd
import csv
```

```python
df = pd.read_csv("base.csv")
df
```

```python
def trans(x):
    if x>0 and x<1000:
        return 1
    elif x>1000 and x<2500:
        return 2
    else:
        return 3
```

```python
df["Ingreso2"]=[trans(x) for x in df["Ingreso"]]
```

```python
gnb = GaussianNB()
modelo=gnb.fit(df[['Edad','AnosEst']],df['Ingreso2'])
```

```python
modelo.predict([50,25])
```
