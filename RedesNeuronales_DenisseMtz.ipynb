```python
import pybrain
from pybrain.tools.shortcuts import buildNetwork
from pybrain.datasets import SupervisedDataSet
from pybrain.supervised.trainers import BackpropTrainer

import pandas as pd
import numpy as np
import csv
```

```python
rn = pd.read_csv("BaseRN2.csv")
rn
```

```python
net = buildNetwork(4,8,1)
```

```python
ds = SupervisedDataSet(4, 1) #tres var dep y una ind
ds.setField('input',rn[['GENDER','AGE','HEIGHT','WEIGHT']])
ds.setField('target',rn[['VICTIM MINOR']])
```

```python
trainer = BackpropTrainer(net, ds)
for i in range(10000):
    trainer.train()
```

```python
r1 = net.activate([0,46,507,200]) #debe dar 0
r2 = net.activate([0,61,509,160]) #debe dar 1

print(r1),
print(r2)
```
